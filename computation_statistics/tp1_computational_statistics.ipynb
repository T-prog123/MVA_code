### Question 1

We have $X_1 \sim \mathrm{Unif}(0,\theta)$ with density $f(x) = \frac{1}{\theta}\mathbf{1}_{[0,\theta]}(x)$.

$$
\mathbb{E}_\theta(X_1)
= \int_0^\theta x \cdot \frac{1}{\theta}\,dx
= \frac{1}{\theta}\left[\frac{x^2}{2}\right]_{0}^{\theta}
= \frac{\theta}{2}.
$$

Hence, for the sample mean $\bar{X} = \frac{1}{n}\sum_{i=1}^n X_i$,  

$$
\boxed{\hat{\theta}_1 = 2\bar{X}}
$$

is unbiased for $\theta$.

### Question 2

$\hat{\theta}_1$ is unbiased so:

$$
R_\theta(\hat{\theta}_1) = \mathrm{MSE}(\hat{\theta}_1) = \mathrm{Var}(\hat{\theta}_1).
$$

For one observation $X_i \sim \mathrm{Unif}(0,\theta)$:

$$
\mathrm{Var}(X_i)
= \mathbb{E}[X_i^2] - (\mathbb{E}[X_i])^2
= \left(\int_0^\theta x^2 \frac{1}{\theta} dx\right) - \left(\frac{\theta}{2}\right)^2
= \frac{1}{\theta}\left[\frac{x^3}{3}\right]_0^\theta - \frac{\theta^2}{4}
= \frac{\theta^2}{3} - \frac{\theta^2}{4}
= \frac{\theta^2}{12}.
$$

Then

$$
\mathrm{Var}(\bar{X}) = \frac{1}{n}\mathrm{Var}(X_1) = \frac{\theta^2}{12n}, \qquad
\mathrm{Var}(\hat{\theta}_1) = 4\,\mathrm{Var}(\bar{X}) = \frac{\theta^2}{3n}.
$$

So the quadratic risk is

$$
\boxed{R_\theta(\hat{\theta}_1) = \frac{\theta^2}{3n}}.
$$

### Question 3

The likelihood is

$$
L(\theta; x_{1:n})
= \prod_{i=1}^n \frac{1}{\theta}\,\mathbf{1}_{[0,\theta]}(x_i)
= \theta^{-n}\,\mathbf{1}\{\theta \ge \max_i x_i\}.
$$

For $\theta \ge \max_i x_i$, $L(\theta) = \theta^{-n}$ is **decreasing** in $\theta$.  
Hence, the maximum occurs at the smallest admissible value of $\theta$:

$$
\boxed{\hat{\theta}_2 = \max_{1 \le i \le n} X_i.}
$$

### Question 4

Let $Y = \max_i X_i$. For $0 \le y \le \theta$:

$$
F_Y(y) = \Pr(Y \le y) = \Pr(X_1 \le y, \dots, X_n \le y)
= \left(\frac{y}{\theta}\right)^n,
$$

$$
f_Y(y) = F_Y'(y) = \frac{n}{\theta^n} y^{n-1}.
$$

**Expectation:**

$$
\mathbb{E}_\theta[Y]
= \int_0^\theta y f_Y(y)\,dy
= \int_0^\theta y \cdot \frac{n}{\theta^n} y^{n-1} dy
= \frac{n}{\theta^n} \int_0^\theta y^n dy
= \frac{n}{\theta^n} \left[\frac{y^{n+1}}{n+1}\right]_0^\theta
= \frac{n}{n+1}\,\theta.
$$

**Second moment:**

$$
\mathbb{E}_\theta[Y^2]
= \frac{n}{\theta^n} \int_0^\theta y^{n+1} dy
= \frac{n}{\theta^n} \left[\frac{y^{n+2}}{n+2}\right]_0^\theta
= \frac{n}{n+2}\,\theta^2.
$$

**Variance and bias:**

$$
\mathrm{Var}_\theta(Y)
= \mathbb{E}[Y^2] - (\mathbb{E}[Y])^2
= \frac{n}{n+2}\theta^2 - \left(\frac{n}{n+1}\theta\right)^2
= \frac{n}{(n+1)^2(n+2)}\,\theta^2,
$$

$$
\mathrm{Bias}_\theta(Y)
= \mathbb{E}[Y] - \theta
= -\frac{\theta}{n+1}
$$

**Quadratic risk:**

$$
R_\theta(\hat{\theta}_2)
= \mathrm{Var}(Y) + \mathrm{Bias}^2
= \frac{n}{(n+1)^2(n+2)}\theta^2 + \frac{1}{(n+1)^2}\theta^2
= \frac{2}{(n+1)(n+2)}\,\theta^2.
$$

$$
\boxed{R_\theta(\hat{\theta}_2) = \frac{2\,\theta^2}{(n+1)(n+2)}}.
$$


### Question 5

We compare the two risks:

$$
R_\theta(\hat{\theta}_1) = \frac{\theta^2}{3n}, \qquad
R_\theta(\hat{\theta}_2) = \frac{2\theta^2}{(n+1)(n+2)}.
$$

We test when $R_\theta(\hat{\theta}_2) \le R_\theta(\hat{\theta}_1)$:

$$
\frac{2}{(n+1)(n+2)} \le \frac{1}{3n}
\Longleftrightarrow
6n \le (n+1)(n+2)
\Longleftrightarrow
0 \le n^2 - 3n + 2
\Longleftrightarrow
0 \le (n-1)(n-2).
$$

Hence:
- For $n = 1, 2$: equal risk.  
- For $n \ge 3$: $R_\theta(\hat{\theta}_2) < R_\theta(\hat{\theta}_1)$.

$$
\boxed{\text{Therefore, }\hat{\theta}_2  \text{ is the better estimator.}}
$$

### Question 1

For any bounded measurable function $g:\mathbb R^2\to\mathbb R$ we have

$$\mathbb E[g(X,Y)] = \int_0^{2\pi}\int_0^\infty g(r\cos\theta,r\sin\theta)\frac{1}{2\pi}r e^{-r^2/2}\,dr\,d\theta.$$

Using the change of variables $x=r\cos\theta$ and $y=r\sin\theta$, we have $dx\,dy=r\,dr\,d\theta$.

Thus,

$$\mathbb E[g(X,Y)] = \frac{1}{2\pi}\int_{\mathbb R^2} g(x,y)e^{-(x^2+y^2)/2}\,dx\,dy.$$

Therefore, the joint density of $(X,Y)$ is

$$p_{X,Y}(x,y)=\frac{1}{2\pi}\exp\big(-\tfrac{1}{2}(x^2+y^2)\big)
=\frac{1}{\sqrt{2\pi}}e^{-x^2/2}\frac{1}{\sqrt{2\pi}}e^{-y^2/2}.$$

Hence $X\sim\mathcal N(0,1)$, $Y\sim\mathcal N(0,1)$ and they are independent.


### Question 2

Let $Z$ follow a Rayleigh distribution with density $f_Z(r)=r e^{-r^2/2}$ for $r>0$.

For $x\ge0$,

$$P(Z\le x)=\int_0^x r e^{-r^2/2}\,dr = 1 - e^{-x^2/2}.$$

Hence the cumulative distribution function is $F(x)=1-e^{-x^2/2}$.

Let $y=F(x)$, then $1-y=e^{-x^2/2}$ which gives $x^2=-2\ln(1-y)$, so

$$x = \sqrt{-2\ln(1-y)} = F^{-1}(y).$$

By the inverse transform method, if $U\sim U(0,1)$, we can sample

$$Z = F^{-1}(U) = \sqrt{-2\ln(1-U)}.$$

Since $\Theta\sim U(0,2\pi)$, we can then set $X=R\cos\Theta$ and obtain a sampling of $X\sim\mathcal N(0,1).$

```python
n_samples = 100
dimension = 2
# we generate the data
X_base = np.random.uniform(0,1, (n_samples, dimension))
# we add the one column: this is the data we will work on from now on
X = np.concatenate((X_base, np.ones((n_samples,1))), axis=1)

w = np.array([1,2,-1])
w =  w / np.linalg.norm(w)
Y = X @ w
positive_points = Y >= 0 # mask of positive points

def line_function(x, w):
    assert len(w) == 3
    assert w[1] != 0
    return (1/w[1]) * (-w[0] * x + - w[2])

plt.scatter(X_base[positive_points,0], X_base[positive_points,1], color="blue", label="positive")
plt.scatter(X_base[~positive_points,0], X_base[~positive_points,1], color="red", label="negative")
plt.plot(np.linspace(0, 1, 100), line_function(np.linspace(0, 1, 100), w), color="black", label="seperator line", lw = 0.5)
plt.title("Scatter plot of the data")
plt.legend()
plt.xlabel("feature 1")
plt.ylabel("feature 2")
```

### Question 2

We notice that we are in the classical Ordinary Least Squares (OLS) framework of linear regression,  
applied here for classification. 
The empirical risk is defined as:

$$
L_n(\bar{w}) = \frac{1}{n} \| Y - X\bar{w} \|^2
$$

where  
- $X \in \mathbb{R}^{n \times d}$ is the design matrix (each row is $\bar{x}_i = (x_i, 1)$),  
- $Y \in \mathbb{R}^n$ is the vector of labels $y_i \in \{-1, +1\}$,  
- $\bar{w} \in \mathbb{R}^{d}$ is the parameter vector (including bias).

Expanding the squared norm:

$$
L_n(\bar{w}) = \frac{1}{n} (Y - X\bar{w})^\top (Y - X\bar{w})
$$

The gradient with respect to $\bar{w}$ is:

$$
\nabla_{\bar{w}} L_n(\bar{w}) = -\frac{2}{n} X^\top (Y - X\bar{w})
$$

**Stochastic Gradient Descent (SGD) Algorithm Description**

1. **Initialization:**  
   Choose an initial vector $\bar{w}^{(0)}$ (often zero or small random values)  
   and a learning rate $\text{lr} > 0$.

2. **Iterative update:**  
   For each sample $(\bar{x}_i, y_i)$, update the parameter using the stochastic gradient:
   $$
   \bar{w}^{(t+1)} = \bar{w}^{(t)} - \text{lr} \, \nabla_{\bar{w}} \ell_i(\bar{w}^{(t)}),
   $$
   where
   $$
   \ell_i(\bar{w}) = (y_i - \bar{x}_i^\top \bar{w})^2,
   $$
   and
   $$
   \nabla_{\bar{w}} \ell_i(\bar{w}) = -2 \bar{x}_i (y_i - \bar{x}_i^\top \bar{w}).
   $$

3. **Repeat:**  
   Loop over the dataset a fixed number of epochs


### Question 3

```python
# we define the functon
def SGD(X, Y, lr=0.01, epoch=3, w0 = np.array([1,1,0], dtype=float)):
    w = w0.copy()
    for i_epoch in range(epoch + 1):
        for index, x_i in enumerate(X):
            # compute the gradient and update w:
            y_i = Y[index]
            grad = -2 * (y_i - x_i.T @ w) * x_i
            w -= lr * grad
    
    # return the normalied vector
    return w / np.linalg.norm(w) 


def accuracy(w_estimate, X, positive_points):
    positive_points_estimates = X @ w_estimate >= 0
    correct_estimates = sum(positive_points_estimates == positive_points)
    return float(correct_estimates / X.shape[0])
```

```python
# We test on one example

learning_rate = 1 / n_samples
w_estimate =  SGD(X, Y, lr= learning_rate)
mse_w = np.mean((w - w_estimate)**2)
accuracy_w = accuracy(w_estimate, X, positive_points)
cos_sim_w = (w @ w_estimate) / (np.linalg.norm(w)*np.linalg.norm(w_estimate))
print(f"The estimate of W through SGD was {w_estimate}, the true value was {w}")
print(f"That is an MSE of {mse_w}, a cosine similarity of {cos_sim_w} and an accuracy of {accuracy_w}")

plt.scatter(X_base[positive_points,0], X_base[positive_points,1], color="blue", label="positive")
plt.scatter(X_base[~positive_points,0], X_base[~positive_points,1], color="red", label="negative")
plt.plot(np.linspace(0, 1, 100), line_function(np.linspace(0, 1, 100), w_estimate), color="black", label="estimator seperator line", lw = 0.5)
plt.title("Scatter plot of the data")
plt.legend()
plt.xlabel("feature 1")
plt.ylabel("feature 2")
```

```python
# We further test to see the influence of the learning rate and the w0 initialisation

# Choose 3 initialisations and 3 learning rates
w0_list = [
    np.array([0.0, 0.0, 0.0], dtype=float),
    np.array([1.0, 1.0, 0.0], dtype=float),
    np.array([-0.5, 0.8, 0.2], dtype=float),
]
lr_base = 1.0 / len(Y)
lr_list = [0.2 * lr_base, 1.0 * lr_base, 5.0 * lr_base]

fig, axes = plt.subplots(3, 3, figsize=(10, 10), sharex=True, sharey=True)
fig.suptitle("Effect of init $\\bar w^{(0)}$ and learning rate on Adaline (SGD)")

for r, w0 in enumerate(w0_list):
    for c, lr in enumerate(lr_list):
        ax = axes[r, c]
        # fit
        w_estimate = SGD(X, Y, lr=lr, epoch=10, w0=w0)
        # metrics
        mse_w = np.mean((w - w_estimate)**2)
        cos_sim = (w @ w_estimate) / (np.linalg.norm(w) * np.linalg.norm(w_estimate))
        accuracy_w = accuracy(w_estimate, X, positive_points)

        # scatter once per panel
        ax.scatter(X_base[positive_points, 0], X_base[positive_points, 1], color="blue", s=8, label="positive")
        ax.scatter(X_base[~positive_points, 0], X_base[~positive_points, 1], color="red", s=8, label="negative")

        # separator
        xs = np.linspace(0, 1, 100)
        ys = line_function(xs, w_estimate)
        ax.plot(xs, ys, color="black", lw=0.5, label="estimated separator")
        ax.set_title(f"lr = {lr:.3g},  $\\bar w^{{(0)}}$ = {np.array2string(w0, precision=2)}", fontsize=9)

        ax.set_xlim(0, 1); ax.set_ylim(0, 1)

        # metrics in-plot
        ax.text(0.02, 0.97,
                f"cos = {cos_sim:.3f}\nMSE = {mse_w:.3f}\nAccuracy={accuracy_w:.3f}",
                transform=ax.transAxes, va="top", ha="left",
                bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="0.7", lw=0.5, alpha=0.9),
                fontsize=9)

# one legend for the whole figure
handles, labels = axes[0,0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", ncol=3, frameon=False)
plt.tight_layout(rect=[0, 0.05, 1, 0.97])
plt.show()
```

In most configurations (except $w_{0} = 0$), we see that $\hat{w}$ is close to the true $w$. We conclude:
